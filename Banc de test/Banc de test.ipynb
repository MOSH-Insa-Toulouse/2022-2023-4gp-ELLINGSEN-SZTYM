```python
import matplotlib.pyplot as plt
import numpy as np

#Pour le HB en tension

#Mesure de la résistance et calcul de DeltaR/R en %
R0 = 340.0
R = [340.0,294.0,290.0,286.0,282.0,272.0,268.0,260.0]
DeltaRdivR = [0,100*46/R0,100*50/R0,100*54/R0,100*58/R0,100*68/R0,100*72/R0,100*80/R0]

#Calcul de la déformation 
#Cercle : De 5 à 2 cm par pas de 0.5 cm
#deformation = e/(2* Rayon_Cercle)
e=0.3e-3
deformation = [0,e/(2*5.0e-2),e/(2*4.5e-2),e/(2*4.0e-2),e/(2*3.5e-2),e/(2*3.0e-2),e/(2*2.5e-2),e/(2*2.0e-2)] 

#Calcul des coefficients de la droite 
coeff = np.polyfit(deformation,DeltaRdivR,1)
a = coeff[0]
b = coeff[1]
print(a)
print(b)

plt.figure(figsize=(10, 10))
plt.xlabel("Deformation")
plt.ylabel("DeltaR/R in %")
plt.title("HB Pencil in tension")
plt.plot(deformation, DeltaRdivR,"x")

X = np.array(deformation) 
plt.plot(X, a*X+b,"-")

plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#Pour le HB en tension
R0 = 340.0
R = [340.0,294.0,290.0,286.0,282.0,272.0,268.0,260.0]
DeltaRdivR_HBtension = [0,100*46/R0,100*50/R0,100*54/R0,100*58/R0,100*68/R0,100*72/R0,100*80/R0]
e=0.3e-3
deformation = [0,e/(2*5.0e-2),e/(2*4.5e-2),e/(2*4.0e-2),e/(2*3.5e-2),e/(2*3.0e-2),e/(2*2.5e-2),e/(2*2.0e-2)] 

#Calcul des coefficients de la droite 
coeff = np.polyfit(deformation,DeltaRdivR_HBtension,1)
a_HBtension = coeff[0]
b_HBtension = coeff[1]
print(a_HBtension)
print(b_HBtension)

#Pour le B en tension
R0 = 690.0
R = [690.0,680.0,685.0,670.0,650.0,645.0,630.0,610.0]
DeltaRdivR_Btension = [0,100*10/R0,15*50/R0,20*54/R0,40*58/R0,100*45/R0,100*60/R0,100*80/R0]

#Calcul des coefficients de la droite 
coeff = np.polyfit(deformation,DeltaRdivR_Btension,1)
a_Btension = coeff[0]
b_Btension = coeff[1]
print(a_Btension)
print(b_Btension)

plt.figure(figsize=(10, 10))
plt.xlabel("Deformation")
plt.ylabel("DeltaR/R in %")
plt.title("HB and B pencil in tension")
plt.plot(deformation, DeltaRdivR_HBtension, label="HB pencil", color="r", linestyle="", marker="x")
plt.plot(deformation, DeltaRdivR_Btension, label="B pencil", color="b", linestyle="", marker="x")
plt.legend()


X = np.array(deformation) 
plt.plot(X, a_HBtension*X+b_HBtension,"r-")
X = np.array(deformation) 
plt.plot(X, a_Btension*X+b_Btension,"b-")

plt.grid()
plt.show()
```
